```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Microsoft_Stock.csv')
```

```python
df2 = pd.read_csv('Microsoft_Stock_Test.csv')
```

```python
df = df.set_index('Date')
```

```python
df2 = df2.set_index('Date')
```

```python
df.index = pd.to_datetime(df.index).date
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df2.index = pd.to_datetime(df2.index).date
```

```python
df2.index = pd.to_datetime(df2.index)
```

```python
df = df.resample('D').ffill()
```

```python
df2 = df2.resample('D').ffill()
```

```python
df.head(15)
```

```python
df2.head(10)
```

```python
df = df.reset_index().rename(columns={"index":"Date"})
```

```python
df2 = df2.reset_index().rename(columns={"index":"Date"})
```

```python
df['Average'] = ((df['High'] + df['Low']) / 2.0).round(2)
```

```python
df2['Average'] = ((df2['High'] + df2['Low']) / 2.0).round(2)
```

```python
df
```

```python
df['Net'] = (df['Close'] - df['Open']).round(2)
```

```python
df
```

```python
df.to_csv('Microsoft_Stock_Clean.csv', index = False)
```

```python
df2.to_csv('Microsoft_Stock_Test_Clean.csv', index = False)
```
