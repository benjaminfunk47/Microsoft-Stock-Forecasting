```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Microsoft_Stock.csv')
```

```python
df
```

```python
df.dtypes
```

```python
df = df.set_index('Date')
```

```python
df.index = pd.to_datetime(df.index).date
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df = df.resample('D').ffill()
```

```python
df.head(15)
```

```python
df = df.reset_index().rename(columns={"index":"Date"})	
```

```python
df['Average'] = ((df['High'] + df['Low']) / 2.0).round(2)
```

```python
df
```

```python
df['Net'] = (df['Close'] - df['Open']).round(2)
```

```python
df
```

```python
df['Day_of_Week'] = df['Date'].dt.day_name()
mapping = {
    'Sunday': 0,
    'Monday': 1,
    'Tuesday': 2,
    'Wednesday': 3,
    'Thursday': 4,
    'Friday': 5,
    'Saturday': 6
}
df['Day_of_Week'] = df['Day_of_Week'].replace(mapping)
```

```python
df.head(15)
```

```python
#df['Stock_ID'] = 1
df.to_csv('Microsoft_Stock_Clean.csv', index = False)
```
